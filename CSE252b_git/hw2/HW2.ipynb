```python
# Note that np.sinc is different than defined in class
import math
def Sinc(x):
    # Returns a scalar valued sinc value
    """your code here"""
    if x == 0:
        y = 1
    else:
        y = math.sin(x)/x
    return y

def Jacobian(P,p,X):
    # compute the jacobian matrix
    #
    # Input:
    #    P - 3x4 projection matrix
    #    p - 11x1 homogeneous parameterization of P
    #    X - 3xn 3D scene points
    # Output:
    #    J - 2nx11 jacobian matrix
    J = np.zeros((2*X.shape[1],11))
    
    """your code here"""
    X = Homogenize(X)
    x = np.dot(P,X)
    x_de = Dehomogenize(x)
    p_bar = P.reshape(12,1)
    for i in range(X.shape[1]):
        pt_cur = X[:,i]
        w = np.dot(P[-1,:],pt_cur)
        dx_dp = np.zeros((2,12))
        dx_dp[0,:4] = dx_dp[1,4:8] = pt_cur.T
        dx_dp[0,8:] = -x_de[0][i] * pt_cur.T
        dx_dp[1,8:] = -x_de[1][i] * pt_cur.T
        dx_dp = (1/w)* dx_dp
        
        a = p_bar[0,0]
        b = p_bar[1:,0]
        norm_p = np.linalg.norm(p)
        if  norm_p == 0:
            da_dv = np.zeros((1,11))
            db_dv = 0.5*np.eye(11)
        else:
            da_dv = -0.5 * b.T
            half_norm = norm_p / 2
            tmp = math.cos(half_norm)/ half_norm - 
            math.sin(half_norm)/ (half_norm**2)
            db_dv = Sinc(half_norm)* np.eye(11)/ 2 + 
            1/(4*norm_p)*tmp * np.dot(p,p.T)
        dp_dp = np.vstack((da_dv,db_dv))
        Ai = np.dot(dx_dp,dp_dp)
        J[2*i:2*i+2,:] = Ai
    return J

def Parameterize(P):
    # wrapper function to interface with LM
    # takes all optimization variables and parameterizes all of them
    # in this case it is just P, but in future assignments it will
    # be more useful
    #print("shape of P:{}".format(P.shape))
    return ParameterizeHomog(P.reshape(-1,1))

def Deparameterize(p):
    # Deparameterize all optimization variables
    return DeParameterizeHomog(p).reshape(3,4)

def ParameterizeHomog(V):
    # Given a homogeneous vector V return its minimal parameterization
    """your code here"""
    v = V / np.linalg.norm(V)
    #print("shape of v: {},v:{}".format(v.shape,v))
    a = v[0][0]
    b = v[1:,:]
    v_hat = 2/(Sinc(math.acos(a)))*b
    v_norm = np.linalg.norm(v_hat)
    if v_norm > math.pi:
        scale = 1-(2*math.pi/v_norm)*
                math.ceil((v_norm - math.pi)/(2*math.pi))
        v_hat = scale * v_hat
    return v_hat

def DeParameterizeHomog(v):
    # Given a parameterized homogeneous vector return its deparameterization
    """your code here"""
    v_norm = np.linalg.norm(v)
    a = math.cos(v_norm/2)
    b = Sinc(v_norm /2 )*0.5 * v.T
    v_bar = np.zeros((v.shape[0]+1,1))
    v_bar[0,0] = a
    v_bar[1:v.shape[0]+1,:] = b.T
    v_bar = v_bar/ np.linalg.norm(v_bar)
    return v_bar
```
